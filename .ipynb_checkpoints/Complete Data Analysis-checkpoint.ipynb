### Annual Income and GDP

```python
income_gdp= df.drop(['corruption_index','unemployment_rate'], axis=1)
print(income_gdp)
```

```python
# line 1 points
x1 = income_gdp['annual_income']
y1 = income_gdp['country']
# plotting the line 1 points 
plt.plot(x1, y1, label = "annual_income")
  
# line 2 points
x2 = income_gdp['gdp_per_capita']
y2 = income_gdp['country']
# plotting the line 2 points 
plt.plot(x2, y2, label = "gdp_per_capita")
  
# naming the x axis
plt.xlabel('Rate of Annual Income & GDP')
# naming the y axis
plt.ylabel('Country')
# giving a title to my graph
plt.title('Annual Income and GDP relation graph!')
  
# show a legend on the plot
plt.legend()
 
# function to show the plot
plt.show()
```

### Unemployment EDA

```python
unemployment_df1= df.drop(['annual_income','corruption_index','gdp_per_capita'], axis=1)
print(unemployment_df1)

unemployment_df = df.sort_values(by='unemployment_rate',ascending=False)
unemployment_df.plot(x='country', y='unemployment_rate', rot=90, kind='bar', figsize=(20,10))
```

### Unemployment Rate and Corruption Rate

```python
unemployment_corruption= df.drop(['annual_income','gdp_per_capita'], axis=1)
print(unemployment_corruption)
```

```python
unemployment_corruption['corruption_index'].plot(kind='bar', color='seagreen')
unemployment_corruption['corruption_index'].plot(kind='line', color='lime', ms=1)
unemployment_corruption['unemployment_rate'].plot(kind='line', marker='*', color='midnightblue', ms=10)

plt.legend()

plt.show()
```
